# ggplot notes python 3/21

```python
import pandas as pd
from ggplot import * #import all of ggplot
```

```python
%matplotlib inline
```

```python
surveys_complete = pd.read_csv('surveys_complete.csv', index_col=0)
```

```python
surveys_complete.head(10)
```

```python
ggplot(aesthetics= aes(x = 'weight', y = 'hindfoot_length'), data = surveys_complete) #added aesthetics using aes, this is just the basic plot without the datatype plotted
```

```python
ggplot(aes(x = 'weight', y = 'hindfoot_length'), data = surveys_complete) + geom_point() #this is what adds the data to the plot
```

```python
surveys_plot = ggplot(aes(x = 'weight', y = 'hindfoot_length'), data = surveys_complete)
surveys_plot + geom_point() #this code accomplishes the same function as above
```

```python
ggplot(aes(x = 'weight', y = 'hindfoot_length'), data = surveys_complete) + \
    geom_point(alpha = 0.1)
```

```python
ggplot(aes(x = 'weight', y = 'hindfoot_length'), data = surveys_complete) + \
    geom_point(alpha = 0.1)
```

```python
ggplot(aes(x = 'weight', y = 'hindfoot_length', color='species_id'),data = surveys_complete) + \
    geom_point( alpha = 0.1)
```

```python
ggplot(aes(x = 'weight', y = 'hindfoot_length', color = 'sex'),data = surveys_complete[surveys_complete.species_id == 'RM']) + \
    geom_point()
```

```python
ggplot( aes(x = 'species_id', y = 'hindfoot_length'), data = surveys_complete) + geom_boxplot()
```

```python
ggplot(aes(x = 'species_id', y = 'weight'), data = surveys_complete) + geom_violin() + scale_y_log(base=10)
```

```python
ggplot(aes(x = 'weight'), data=surveys_complete) + geom_histogram(binwidth=2)
```

```python
ggplot(aes(x = 'weight', fill='species_id'), data=surveys_complete) + geom_histogram(binwidth=2) #fill changes inside the bars
```

```python
ggplot(aes(x = 'weight', color ='species_id'), data=surveys_complete) + geom_histogram(binwidth=2) #color changes line outside the bars
```
